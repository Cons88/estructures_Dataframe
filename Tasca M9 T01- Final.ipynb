```python
import string 

#string.puntuation contains all the common puntuations
# ans this function will convert each on into "nothing"

def strip_punctuations(line):
    for character in string.punctuation:
        line = line.replace(character,"")
        return line
    
```

```python
# This Dataset consists dataset of Twitter messages labeled with 3 sentiments: 0 (negative), 2 (neutral), and 4 (positive)
with open("C:\\Users\\flore\\OneDrive\\Escritorio\\Texto.txt") as fh:
    filepath = fh.read()
```

**Word frequency**

```python
# Word tokenization
tokenized_book = word_tokenize(filepath)
```

```python
# Remove punctuation
tokenized_book = [word for word in tokenized_book if word.isalnum()]
```

```python
# Frequency
fdist = FreqDist(tokenized_book)
```

```python
# 30 most common words
plt.figure(figsize=(10,7))
fdist.plot(30, cumulative=False)
plt.show()
```

**Stop words**

```python
# Stop words
stop_words = set(stopwords.words("english"))
```

```python
# Remove stop words
filtered_book = [w for w in tokenized_book  if w not in stop_words]
```

```python
# Word frequency
fdist = FreqDist(filtered_book)
```

```python
# 30 most common words
plt.figure(figsize=(10,7))
fdist.plot(30, cumulative=False)
plt.show()
```

**Stemming**

```python
# Stemming
ps = PorterStemmer()
```

```python
# Stemmed book
stemmed_book=[ps.stem(w) for w in filtered_book]
```

```python
# Word frequency
fdist = FreqDist(stemmed_book)
```

```python
# 30 most common words
plt.figure(figsize=(10,7))
fdist.plot(30, cumulative=False)
plt.show()
```
