```python
movie_headers = ['movie_id', 'Titulo', 'Genero']
```

```python
movies = pd.read_csv('Libro2.csv', sep = '::', engine='python',header = None, names = movie_headers)
movies
```

```python
movies
```

```python
# Aqui he logrado separar la columna Título en dos: Nombre de pelicula y estreno
movies[['Nombre_pelicula','Estreno']] = movies['Titulo'].str.split('(', n=1,expand = True)
movies
```

```python
movies['Estreno'] = movies['Estreno'].str.replace(')', '')
movies
```

```python
movies[['Gen1','Gen2','Gen3']] = movies['Genero'].str.split('|', n=2, expand=True)
movies
```

```python
movies
```

```python
movies_final = movies.drop(['Titulo', 'Genero', 'Gen2', 'Gen3'], axis = 1)
```
