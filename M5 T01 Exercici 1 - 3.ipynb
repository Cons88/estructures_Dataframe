- Exercici 1
Descarrega el dataset adjunt de dades oficials de la UEFA i selecciona un atribut del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%. 


Data set amb el que vinc treballant: "Lligues europees UEFA". La meva hipòtesi nula serà que la media de partits guanyats pel pais "ENGLAND" ("Country") es la mateixa media que la dels partits guanyats per tots els paisos.

La meva hipòtesi alternativa es que la media dels partits guanyats ("W") per "ENGLAND" es diferent a la media del total de partits guanyats per tots els paisos.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scipy
from scipy import stats
from scipy.stats import ttest_ind
```

```python
df = pd.read_excel("Lligues europees UEFA.xlsx")
df
```

```python
# Filtrar el grup que farè servir a l´hipotesi: England

England_df = df[(df['Country'] == 'ENG')]
England_df
```

```python
England_df['W'].isnull().sum()
```

```python
England_df['W'].mean()
```

```python
England_df['W'].describe()
```

```python
stat, p = stats.ttest_ind(England_df['W'], df['W'])
print('stat =%.3f, p =%.5f' % (stat, p))
if p > 0.05:
    print ('No podem rebutjar hipòtesi nula')
else:
    print('Rebutgem hipòtesi nula')
```

Conclusió: En aquest primer cas el p-value es major que l´alpha 5% o nivel de confiança; aixó vol dir que no podem rebutjar la meva hipòtesi nula que afirma que la media de partits guanyats pel pais "ENGLAND" ("Country") es la mateixa media que la dels partits guanyats per tots els paisos.

Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%.


Els atributs que faré servir seràn: "Country" i "Pts". 
    La Hipotesi 0: Es que el pais "ENGLAND" te una media de punts IGUAL que el pais de "FRANCE"
    La Hipotesi 1: El pais "ENGLAND" te una media de punts DIFERENT a la media de punts del pais de "FRANCE"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scipy
from scipy import stats
from scipy.stats import ttest_ind
```

```python
df = pd.read_excel("Lligues europees UEFA.xlsx")
df
```

Crearé el dataframe del pais "ENGLAND" amb l´ atribut "Pts"

```python
England_df_pts = df[(df['Country'] == 'ENG')]
England_df_pts
```

Crearé el dataframe del pais "FRANCE" amb l´ atribut "Pts"

```python
France_df_pts = df[(df['Country'] == 'FRA')]
France_df_pts
```

Confirm que no hi ha Null values a la columna "Pts"

```python
df['Pts'].isnull().sum()
```

T - Student Test 

```python
stat, p = stats.ttest_ind(England_df_pts['Pts'], France_df_pts['Pts'])
print('stat =%.3f, p =%.5f' % (stat, p))
if p > 0.05:
    print ('No podem rebutjar hipòtesi nula')
else:
    print('Rebutgem hipòtesi nula')
```

- Exercici 3
Continua amb el conjunt de dades adjunt i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.


En aquest cas, observaré l'atribut "L" "Partirs perduts" per tres paisos diferents: ENGLAND, FRANCE i GERMANY.

La meva Hipòtesi nula és:

H0 = La media de partits perduts dels tres paisos es igual

La meva Hipòtesi alternativa és:

H1 = La media de partits perduts per cada un dels paisos mencionats és diferent.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scipy
from scipy import stats
from scipy.stats import ttest_ind
```

```python
df = pd.read_excel("Lligues europees UEFA.xlsx")
df
```

Alemanyia te dos registres menys que França i Anglaterra doncs, he fet un "sample" de 15 registres per pais.

```python
England_L_df= df[(df['Country']== "ENG")].sample(15)
England_L_df
```

```python
France_L_df= df[(df['Country']== "FRA")].sample(15)
France_L_df
```

```python
Germany_L_df= df[(df['Country']== "GER")].sample(15)
Germany_L_df
```

```python
from scipy.stats import f_oneway
```

```python
stat, p = f_oneway(England_L_df['L'],France_L_df['L'], Germany_L_df['L'])
```

```python
print('stat =%.3f, p =%.5f' % (stat, p))
if p > 0.05:
    print ('No podem rebutjar hipòtesi nula')
else:
    print('Rebutgem hipòtesi nula')
```

Conclusió: En aquest tercer cas el p-value es major que l´alpha 5% o nivel de confiança; aixó vol dir que la media de partits perduts per cada pais:"ENGLAND", "FRANCE" i "GERMANY" NO ES signifativament DIFERENT. Per aixó: No podem rebutjar hipòtesi nula.
